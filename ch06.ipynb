```python
# 데이터 합치기
# 가로로 합치기
# 중간고사 데이터
test1 = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})
test2 = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                      'midterm' : [70, 83, 65, 95, 80]})

print(test1)
print(test2)
```

```python
# id를 기준으로 합쳐서 total에 할당
# pd.merge()이용하여 합침
# 오른쪽에 입력한 데이터 프레임을 왼쪽 데이터 프레임에 결합하도록 how = 'left'
# 데이터를 합칠 때 기준으로 삼을 변수명을 on에 작성 

total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
# 분석 도전
midwest = pd.read_csv('midwest.csv')
midwest.head()
```

```python
# 전체 인구 대비 미성년 인구 백분율 변수 추가
midwest = midwest.assign(ratio = (midwest['poptotal'] - midwest['popadults']) / midwest['poptotal'])
midwest.head()
```

```python
# 미성년 인구 백율이 가장 높은 상위 5개 country의 미성년 인구 백분율 출력
midwest.sort_values('popadults_rate', ascending = False)[['county', 'ratio']] \
    .head()
```

```python
midwest = midwest.assign( \
    ratio_rate = np.where(midwest['ratio'] >= 0.4, 'large',
    np.where(midwest['ratio'] >= 0.3, 'middle', 'small')))

midwest['ratio_rate'].value_counts()                       
```

```python
midwest.groupby('ratio_rate') \
    .agg(n = ('ratio_rate', 'count'))
```

```python
# 전체 인구 대비 아시아인 인구 백분율 변수 추가
midwest = midwest.assign(popasian_ratio = midwest['popasian'] / midwest['poptotal'])
midwest
```

```python
midwest[['state', 'county', 'popasian_ratio']] \
    .sort_values('popasian_ratio') \
    .head(10)
```

```python
# 변수 생성, 출력 한거번에 
midwest.assign(popasian_ratio = midwest['popasian'] / midwest['poptotal']) \
     [['state', 'county', 'popasian_ratio']] \
    .sort_values('popasian_ratio') \
    .head(10) 
```
