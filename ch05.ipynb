```python
# total 변수가 20을 넘으면 합격, 넘지 못하면 불합격
import numpy as np
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
```

```python
mpg['test'].value_counts() # 연비 합격 빈도표 만들기
```

```python
count_test = mpg['test'].value_counts()
```

```python
# 연비 합격 빈도 막대 그래프 만들기 
count_test.plot.bar()
```

```python
# 축 이름 회전하기
count_test.plot.bar(rot = 0)
```

```python
# 중첩 조건문 
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                        np.where(mpg['total'] >= 20, 'B', 'C'))

mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# 등급 빈도표 알파벳순 정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
mpg['grade2'] = np.where(mpg['total'] >= 30, 'A',
                np.where(mpg['total'] >= 25, 'B',
                np.where(mpg['total'] >= 20, 'C', 'D')))

mpg.head()
```

```python
# 여러 조건 중 하나만 해당할 경우 
mpg['size'] = np.where((mpg['category'] == 'compact') |
                      (mpg['category'] == 'subcompact') |
                      (mpg['category'] == '2seater'),
                      'small', 'large')

mpg['size'].value_counts()
```

```python
# 여러 조건 중 하나만 해당할 경우 
mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']),
                       'small', 'large')

mpg['size'].value_counts()
```

```python
# 분석 도전
df_midwest = pd.read_csv('midwest.csv')
df_midwest.head()
```

```python
df_midwest = df_midwest.rename(columns = {'poptotal': 'total', 'popasian':'asian'})
df_midwest.head()
```

```python
df_midwest['asian_rate'] = df_midwest['asian'] / df_midwest['total']
df_midwest['asian_rate'].plot.hist()
```

```python
df_midwest['size'] = np.where(df_midwest['asian_rate'].mean() < df_midwest['asian_rate'], 'large', 'small')
count_size = df_midwest['size'].value_counts()
count_size
```

```python
count_size.plot.bar(rot = 0)
```
