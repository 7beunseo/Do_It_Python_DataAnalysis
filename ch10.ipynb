```python
# 연설문 불러오기
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
# 한글이 아닌 모든 문자는 공백으로 변경
import re
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
# 명사 추출하기
import konlpy
hannanum = konlpy.tag.Hannanum()

# 명사 추출하기
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd

# 데이터 프레임으로 전환 
df_word = pd.DataFrame({'word':nouns})
df_word
```

```python
# 문자 수 추가 
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어의 사용 빈도를 구하고 빈도순 정렬
df_word = df_word.groupby('word', as_index = False) \
            .agg(n = ('word', 'count')) \
            .sort_values('n', ascending = False)

df_word
```

```python
# 단어 빈도 막대 그래프 ㅋ만들기
# 상위 20개 단어를 추출
top20 = df_word.head(20)
top20
```

```python
# 막대 그래프 만들기
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.family': 'Malgun Gothic',
                     'figure.dpi': '120',
                     'figure.figsize': [6.5, 6]})

# 막대 그래프 만들기
sns.barplot(data = top20, y = 'word', x = 'n')

# 단어별 빈도 그래프
```

```python
# wc 만들기
from wordcloud import WordCloud
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white')
```

```python
# 워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')

#데이터 살표보기
df.info()
```

```python
#댓글이 들어있는 reply 변수
# 불필요한 문자를 제거하여 한글만 남김 
# str의 경우에는 re.sub() 사용
# 데이터 프레임에 담겨있는 변수는 str.replace()를 사용해야 함 
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
df['reply'].head()
```

```python
# Kkma형태소 분석기 사용하여 댓글에서 명사를 추출
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
# 명사 추출 - apply() 활용 
# 행마다 여러 단어가 리스트 자료 구조로 들어감 
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
# 한 행에 한 단어가 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
# 두글자 이상으로 된 단어만 남기고 단어 빈도표 만들기
# 데이터 프레임 만들기
df_word = pd.DataFrame({'word' : nouns})

#글자 수 세기 
df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
#빈도표 만들기
df_word = df_word.groupby('word') \
            .agg(n = ('word', 'count')) \
            .sort_values('n', ascending = False)

df_word
```

```python
# 단어 빈도 막대 그래프 만들기
top20 = df_word.head(20)
top20
```

```python
#가로 세로 크기 조정
plt.rcParams.update({'figure.figsize': [6.5, 5]})

# 막대 그래프 만들기
sns.barplot(data = top20, y = 'word', x = 'n')
```
